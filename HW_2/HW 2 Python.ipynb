```python
''' 1. В качестве входящих данных запросите целое число. Если оно делится на 3,
выведите «foo»; если оно делится на 5, выведите «bar»; если оно делится на оба,
выведите «ham» (а не «foo» или «bar»). '''

number = int(input('Enter an integer: '))

if number%3 == 0 and number%5 == 0:
    print('ham')
elif number%5 == 0:
    print('bar')
elif number%3 == 0:
    print('foo')
else:
    print('No info')
```

```python
''' 2. В качестве входящих данных запросите два числа и выведите какое из них
меньше и какое большее. '''

number_one = int(input('Enter the first number: '))
number_two = int(input('Enter the second number: '))

if(number_one > number_two):
   print(f'''The greater number is: {number_one}
         \nThe lower number is: {number_two}''')
elif (number_one < number_two):
   print(f'''The greater number is: {number_two}
         \nThe lower number is: {number_one}''')
   
```

```python
''' 3. В качестве входящих данных запросите три числа и выведите наименьшее,
среднее и наибольшее. Предположим, все они разные. '''

number_one = int(input('Enter the first number: '))
number_two = int(input('Enter the second number: '))
number_three = int(input('Enter the third number: '))
input_numbers = (number_one, number_two, number_three)

max_number = max(input_numbers)
min_number = min(input_numbers)
avg_number = sorted(input_numbers)[len(input_numbers) // 2]

print(f"Largest: {int(max_number)}")
print(f"Smallest: {int(min_number)}")
print(f"Average: {int(avg_number)}")

```

```python
''' 5. Сыграйте в игру Fizz-Buzz: выведите все числа от 1 до 100; если число делится на
3, вместо числа выведите «fizz». Если оно делится на 5, вместо числа выведите
“buzz”. Если оно делится на оба, выведите «fizz buzz» вместо числа '''"HW 1 Python.ipynb"

# рішення за допомогою "for"

for numbers in range(0,101):
    if numbers%3 == 0 and numbers%5 == 0:
        print('fizz buzz')
    elif numbers%5 == 0:
        print('buzz')
    elif numbers%3 == 0:
        print('fizz')
    else:
        print(numbers) 

    

```

```python
''' 5. Сыграйте в игру Fizz-Buzz: выведите все числа от 1 до 100; если число делится на
3, вместо числа выведите «fizz». Если оно делится на 5, вместо числа выведите
“buzz”. Если оно делится на оба, выведите «fizz buzz» вместо числа '''"HW 1 Python.ipynb"

# рішення за допомогою while

numbers = 0
while numbers < 101:
    if numbers%3 == 0 and numbers%5 == 0:
        print('fizz buzz')
    elif numbers%5 == 0:
        print('buzz')
    elif numbers%3 == 0:
        print('fizz')
    else:
        print(numbers)
    numbers += 1   
```

```python
''' 6. Сыграйте в игру 7-boom: выведите все числа от 1 до 100; если число делится на 7
или содержит цифру 7, выведите “BOOM” вместо числа '''

# вирішила за допомогою for 

for numbers in range(0,101):
    if '7' in str(numbers) or numbers%7 == 0:
        print('BOOM')
    else:
        print(numbers)

```
