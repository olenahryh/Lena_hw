```python
# 3. Посчитайте количество побед каждого участника, принимая за победу 1-е
# место. Сохраните результаты во вторичном DataFrame

winners = lifts[lifts['place'] == '1'].groupby('name')[['place']].count()
print(winners.sort_values('place', ascending=False))
```

```python
#4. Используйте DataFrame из упражнения 3, чтобы показать для каждой
# комбинации подразделения и пола, участника с наибольшим количеством побед
# среди участников, когда-либо участвовавших в этом дивизионе
result = pd.merge(winners, lifts, how = 'inner', on = ['name'])
result = result[result['place_y'].str.isdigit()]
result['place_y'] = result['place_y'].astype('int')
idx = result.groupby(['sex', 'division'])['place_y'].transform(max) == result['place_y']                    #.sort_values(['sex','place_y'], ascending=False)
d = result[idx]
d[['sex', 'division', 'place_y','name']].sort_values(['sex','place_y'], ascending=False)
```

```python
import mysql.connector
import pandas as pd
cnx = mysql.connector.connect(user='root', password='lans123', host='localhost', database='employees')



query = """SELECT emp_no, dept_no, sum_salary
FROM (
SELECT s.emp_no,dept_no,(salary/12) AS month_salary, 
SUM(salary/12) OVER(partition by dept_no ORDER BY salary/12) sum_salary
FROM salaries AS s
INNER JOIN dept_emp AS d
ON s.emp_no = d.emp_no
AND curdate() BETWEEN d.from_date AND d.to_date
WHERE curdate() BETWEEN s.from_date AND s.to_date

) month_salary
WHERE sum_salary <= 500000
ORDER BY dept_no, month_salary;"""


df = pd.read_sql_query(query,cnx)
print(df)
```
