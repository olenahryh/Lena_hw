```python
# 2. Какие страны являются крупными экспортерами кофе?

# Ми бачимо, що найбільшими експортерами є країни: Mexico, Colombia, Guatemala and Brazil                      

coffee['country_of_origin'].value_counts().head(10)
```

```python
# Зробила візуалізацію на попередній код

sns.set_style('whitegrid')
plt.figure(figsize=(13,4))
sns.countplot(x = 'country_of_origin', data = coffee, order = coffee['country_of_origin'].value_counts().index[:10])
```

```python
# 3. Каковы корреляции между различными показателями оценки кофе?

# Я додала до кореляції ті данні, які, на мою думку, є цікавими для порівняння

cof = coffee[['total_cup_points', 'aroma', 'flavor', 'aftertaste', 'acidity', 'sweetness', 'moisture', 'cupper_points',
               'altitude_low_meters', 'altitude_high_meters', 'altitude_mean_meters']]
cof.corr()
```

```python
# Завдяки heatmap ми можемо побачити силу зв'язку між двома колонками. Наприклад, чим більше total_cup_points, тим краще flavor (correlaton 0.87)
# Інший приклад, чим більший рівень вологи, тим МЕНШ вираженний післясмак (correlation -0.18).
# Є данні, які майже не мають ніякого зв'язку між собою (flavor + sweetness = 0.29)

plt.figure(figsize=(11,7))
sns.heatmap(cof.corr(), annot = True, cmap = 'coolwarm')
```

```python
# 4. Как (если есть) влияние цвета зерен на общий сорт кофе?

# Ми бачимо, що колір зерен Green використовується найбільше для КОЖНОГО сорту кави. Цей колір є найпопулярнішим по статистиці.


coffee_species = coffee.groupby(['species', 'color'])[['color']].value_counts()
print(coffee_species.sort_values(ascending=False))
```

```python
# зробила візуалізацію попереднього коду 

sns.countplot(x = 'color', hue = 'species', data = coffee);
```

```python
# 5. Влияет ли страна происхождения на качество кофе?

# За допомогою кодів нижче, ми бачимо, що звісно країна впливає на якість кави. Бачимо, що такі показники, як:
# аромат, смак та післясмак, кращі в країнах: Ethiopia ТА United States.

quality_aroma = coffee[['country_of_origin', 'aroma']].sort_values(['aroma'], ascending=False)
quality_aroma.head(3)
```
